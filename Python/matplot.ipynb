```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Basic plot")
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

# Create a customized line plot
plt.plot(x,y,color="green",linestyle="--",marker="o",linewidth=3,markersize=10)  # Only options "-","--","-.",":"
```

```python
# Multiple Plots
x=[1,2,3,4,5,6]
y1 = [3,5,9,7,6,43]
y2 = [8,7,6,5,4,32]

plt.figure(figsize=(9,5))
plt.subplot(2,1,1)  # 1 row, 2 columns, first subplot
plt.plot(x, y1, color='blue', linestyle='-', marker='o')
plt.title("Plot 1")

plt.subplot(2,1,2)  # 1 row, 2 columns, first subplot
plt.plot(x, y2, color='blue', linestyle='-', marker='o')
plt.title("Plot 2")
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create a bar plot
plt.bar(categories, values, color='darkgreen', width=0.5)
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create a histogram
plt.hist(data,bins=5, color='purple',)
```

```python
# Scatter Plot
# Sample data
x=[1,2,2,4,5]
y=[6,5,3,4,6]

plt.scatter(x,y,marker="o")
```

```python
# Pie Chart
labels=['A','S','F','E']
val=[7,4,2,3]
colors=["red","blue","green","yellow"]
explode=(0.2,1,0.5,0.3)  # Only "0" or "1" values, "0" means no explosion

plt.pie(val,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,explode=explode,startangle=330)
```

```python
## Sales data Visualization
import pandas as pd
sales_data=pd.read_csv("data.csv")
sales_data.head(5)
```

```python
sales_data.info()
```

```python
sales_by_product=sales_data.groupby(["Region","Product"])["Sales"].sum()
print(sales_by_product)
```

```python
sales_by_product.plot(xlabel="Region",ylabel="Sales",kind="bar",title="Sales",color="teal")
```

```python
# Plot sales trend over time
sales_trend=sales_data.groupby("Date")["Sales"].sum().reset_index()
print(sales_trend)
```

```python
plt.plot(sales_trend["Date"],sales_trend["Sales"],marker="o")
```
