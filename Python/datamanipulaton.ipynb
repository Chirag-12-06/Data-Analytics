```python
##Handling Missing Values
df.isnull() # Check for missing values in the DataFrame
df.isnull().any(axis=1) # Check for any missing values in each row
```

```python
df.isnull().any(axis=0) # Check for any missing values in each column
```

```python
df.isnull().sum() # Count the number of missing values in each column
```

```python
##filling missing values
df.fillna(0) # Fill missing values with 0
df
```

```python
##filling missing values with mean
df['New_Sales']=df['Sales'].fillna(df['Sales'].mean()) # Fill missing values in 'Sales' with the mean of the column
df
```

```python
##Renaming Columns
df=df.rename(columns={'Date' : 'Sales_Date'})
df.head()
```

```python
## Changing data types
df['Value']=df['Value'].fillna(df['Value'].mean()).astype(int) # Convert 'Value' column to integer type after filling missing values
df
```

```python
## Apply function to a column
df['New_value']=df['Value'].apply(lambda x:x*2) # Apply a function to double the values in the 'value' column
df.head()
# df.drop('New_value,axis=1,inplace=True')
```

```python
## Data Aggregation and Grouping
df.groupby('Product')['Value'].mean() # Group by 'Product' and calculate the mean of 'Value'
```

```python
df.groupby(['Region','Product'])['Value'].sum() # Group by 'Product' and 'Region' and calculate the sum of 'Value'
```

```python
## Aggregation with multiple functions
df.groupby('Region')['Value'].agg(['sum','count','mean']) # Group by 'Region' and calculate sum, count, and mean of 'Value'
```

```python
## Merging and Joining DataFrames
# Create samples
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value2': [4, 5, 6]})
```

```python
## Merge on a single key
pd.merge(df1,df2,on='key',how='inner') # Merge df1 and df2 on 'key' using an inner join
```

```python
pd.merge(df1,df2,on='key',how='outer') # Merge df1 and df2 on 'key' using an outer join
```

```python
pd.merge(df1,df2,on='key',how='left') # Merge df1 and df2 on 'key' using an left join
```

```python
pd.merge(df1,df2,on='key',how='right') # Merge df1 and df2 on 'key' using an right join
```
